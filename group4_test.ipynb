```python
def add_func(data):
    data = str(data)
    
    if "개포" in data:
        return "개포동"
    elif "청담" in data:
        return "청담동"
    elif "대치" in data:
        return "대치동"
    elif "삼성" in data:
        return "삼성동"
    elif "일원" in data:
        return "일원동"
    elif "논현" in data:
        return "논현동"
    elif "도곡" in data:
        return "도곡동"
    elif "역삼" in data:
        return "역삼동"
    elif "신사" in data:
        return "신사동"
    elif "수서" in data:
        return "수서동"
    elif "세곡" in data:
        return "세곡동"
    elif "자곡" in data:
        return "자곡동"
    elif "율현" in data:
        return "율현동"
    else:
        return "압구정동"
```

```python
df1['단속지역']=df1['단속지역'].apply(add_func)
```
